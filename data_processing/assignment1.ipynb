```python
#SimpleImputer 이용해보기
#from sklearn.impute import SimpleImputer

#중앙값으로 채워보기
#imputer = SimpleImputer(strategy="median")
#imputed_nfl = imputer.fit_transform(nfl)

#could not convert string to float: '2009-09-10' 같은 에러뜸
#그런데 strategy="constant"로 해주면 잘 되긴 함.. 근데 이게 맞나?

#그래서 encoding을 먼저 해주기로 함!
```

```python
#LabelEncoding 시도해보기
from sklearn.preprocessing import LabelEncoder

#nfl copy본 생성
le_nfl = nfl.copy()

#LabelEncoder
le = LabelEncoder()

#fit_transform
for col in le_nfl:
    le_nfl[col] = le.fit_transform(le_nfl[col].astype(str).values)
    
le_nfl
```

```python
#다시 데이터셋 확인해보기
le_nfl.isnull().sum()
```

```python
#그런데 LabelEncoding을 해줬더니 결측값 자체도 하나의 카테고리로 인코딩되었나봄..
#그래서 imputation을 most_frequent 방식으로 한 다음에, 다시 LabelEncoding을 해보기로 함

from sklearn.impute import SimpleImputer

imputer = SimpleImputer(strategy="most_frequent")
imputed_nfl = imputer.fit_transform(nfl)

imputed_nfl
```

```python
#다시 label encoding진행

le2 = LabelEncoder()
le_nfl = nfl.copy()

for col in le_nfl:
    le_nfl[col] = le2.fit_transform(le_nfl[col].astype(str).values)
    
le_nfl.isnull().sum()
```
