```python
import numpy as np
import pandas as pd

train = pd.read_csv('../dataset/house_prices/train.csv')
test = pd.read_csv('../dataset/house_prices/test.csv')
```

```python
from sklearn.impute import SimpleImputer

columnList = train.columns
imputer = SimpleImputer(strategy="most_frequent")
train = pd.DataFrame(imputer.fit_transform(train))
train.columns = columnList
```

```python
# object형을 다 categorical data로 바꿔주기

for col in train.columns:
    if(train[col].dtype == 'O'):
        train[col] = train[col].astype('category').cat.codes
        
train.describe()
```

```python
# 이제 싹 다 정수형으로 바뀜
train.info()
```

```python
from sklearn.model_selection import train_test_split

X = train.drop(['SalePrice'], axis=1)
y = train['SalePrice']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=20)
```

```python
from sklearn.ensemble import RandomForestRegressor

rf = RandomForestRegressor()
rf.fit(X_train, y_train)
```

```python
from sklearn.model_selection import RandomizedSearchCV

hyperparameter_space = {'max_depth':[2,3,4,6,8], 
                        'min_samples_leaf':[2,4,6,8,10],
                        'min_samples_split':[2,3,4,5]}

rs = RandomizedSearchCV(rf, param_distributions=hyperparameter_space,
                        n_iter=10, scoring="accuracy", random_state=0,
                        n_jobs=-1, cv=10, return_train_score=True)
```

```python
rs.fit(X_train, y_train)
```

```python
print("Optimal hyperparameter combination:", rs.best_params_)
```
