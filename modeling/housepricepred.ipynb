### 기본 설정

```python
import numpy as np
import pandas as pd
```

```python
train = pd.read_csv('../dataset/house_prices/train.csv')
test = pd.read_csv('../dataset/house_prices/test.csv')
```

### Data 살펴보기

```python
train.describe()
```

```python
train.info()
```

```python
train.isnull().sum()
```

### Imputation (`most_frequent`)

```python
from sklearn.impute import SimpleImputer

columnList = train.columns
imputer = SimpleImputer(strategy="most_frequent")
train = pd.DataFrame(imputer.fit_transform(train))
train.columns = columnList
```

```python
train.describe()
```

### Encoding

```python
# object형을 다 categorical data로 바꿔주기

for col in train.columns:
    if(train[col].dtype == 'O'):
        train[col] = train[col].astype('category').cat.codes
        
train.describe()
```

```python
# 이제 싹 다 정수형으로 바뀜
train.info()
```

### Train Test Split

```python
from sklearn.model_selection import train_test_split

X = train.drop(['SalePrice'], axis=1)
y = train['SalePrice']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=20)
```

### Decision Tree

```python
from sklearn import tree

dectree = tree.DecisionTreeRegressor()
```

```python
fittree = dectree.fit(X_train, y_train)
```

```python
fittree.score(X_test, y_test)
```

```python
from sklearn.metrics import mean_squared_error

pred = fittree.predict(X_test)
RMSE = mean_squared_error(y_test, pred)**0.5

RMSE
```

```python
# hyperparameter space 정의함
param_space = {
    'max_depth': [2,3,4,6],
    'min_samples_leaf': [1,2,4,6],
    'min_samples_split': [2,3,4,8],
}
```

```python
# GridSearchCV 모듈 이용
from sklearn.model_selection import GridSearchCV
gs = GridSearchCV(fittree, param_grid=param_space, scoring='accuracy', n_jobs=-1, cv=10, return_train_score=True)
```

```python
# model 대신 gs 객체에 fit 함
gs.fit(X_train, y_train)
print("Optimal hyperparameter combination: ", gs.best_params_)
```

```python
print("Mean cross-validated training accuracy score:", gs.best_score_)
```

```python
from sklearn.model_selection import RandomizedSearchCV

hyperparameter_space = {'max_depth':[2,3,4,6,8], 
                        'min_samples_leaf':[2,4,6,8,10],
                        'min_samples_split':[2,3,4,5]}

rs = RandomizedSearchCV(fittree, param_distributions=param_space,
                        n_iter=10, scoring="accuracy", random_state=0,
                        n_jobs=-1, cv=10, return_train_score=True)
```

```python
rs.fit(X_train, y_train)
```

```python
print("Optimal hyperparameter combination:", rs.best_params_)
```
